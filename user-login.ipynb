```python
#--------------------------DATABASE CONNECTION-----------------------------------------

import mysql.connector

conn = mysql.connector.connect(

    host = "localhost",
    user = "root",
    password = "root",
    database = "mydb"
    
)

cursor = conn.cursor()

#-------------------------------------Create class--------------------------------------------

class Login():

    
    def __init__(self):
        self.menu()


    def menu(self):
        user_input = input("""

        1. press 1 for new registration
        2. press 2 for login 
        3. press 3 for exit
        """)

        if user_input == '1':
            self.new_registration()
        elif user_input == '2':
            self.new_login()
        elif user_input == "3":
            print("Exit")
            conn.close()
            exit()
        else:
            print("Invalid choice ❌")

    #-----------------------------------------------NEW REGISTRATION-------------------------------

    def new_registration(self):
        username = input("Enter your id:")
        self.username = username
        password = input("Enter your password:")
        self.password = password

       

        try:
            query = "INSERT INTO users (username, password) VALUES (%s, %s)"
            cursor.execute(query, (username, password))
            conn.commit()
            print("Registration Successful ✅")

        except mysql.connector.IntegrityError:
            print("Username already exists ❌")

        self.menu()

    #----------------------------------LOGIN---------------------------------------

    def new_login(self):
        username = input("Enter your id:")
        password = input("Enter your password:")

        query = "SELECT * FROM users WHERE username=%s AND password=%s"
        cursor.execute(query, (username, password))

        user = cursor.fetchone()

        if user:
            print("Login Successful ✅ Welcome", username)
            self.dashboard(username)
        else:
            print("Invalid Credentials ❌")

        self.menu()

    #-----------------------------------user dashboard----------------------------------------------------

    def dashboard(self, username):
        
        choice = input("""
        1. press 1 to add data 
        2. press 2 to view data
        3. press 3 logout
        """)

        if choice == '1':
            self.add_data(username)
        elif choice == '2':
            self.view_data(username)
        elif choice == "3":
            print("logout")
            self.menu
            
        else:
            print("Invalid choice ❌")

    #------------------------------------------add data ito username-------------------------------

    def add_data(self, username):
        name = input("Enter Name: ")
        surname = input("Enter Surname: ")
        email = input("Enter Email: ")
        mobile = input("Enter Mobile Number: ")
        age = input("Enter Age: ")

        query = """ INSERT INTO profile(username, name, surname, email, mobile, age)
                    VALUES (%s, %s, %s, %s, %s, %s) """

        cursor.execute(query,(username, name, surname, email, mobile, age))

        conn.commit()

        print("Profile Saved Successfully ✅")

#---------------------------------- view data------------------------------------

    def view_data(self, username):

        query = """ SELECT name,surname, email, mobile, age FROM profile WHERE username = %s """

        cursor.execute(query,(username,))

        records = cursor.fetchall()

        if records:
            
            for row in records:
                print("\n----- Profile -----")
                print("Name:", row[0])
                print("Surname:", row[1])
                print("Email:", row[2])
                print("Mobile:", row[3])
                print("Age:", row[4])

        else:
            print("No Data Found")
        
        self.dashboard(username)
        
# run below commnd to run the program

obj = Login()
```
